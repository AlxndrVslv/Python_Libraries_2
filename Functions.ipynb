```python
import pandas as pd
```

```python
movies = pd.read_csv('ml-latest-small/movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

# Задание 1

```python
def rate_level(rating):
    if rating <= 2: return 'low'
    elif rating <= 4: return 'middle'
    else: return 'high'
```

```python
ratings['class'] = ratings['rating'].apply(rate_level)
print(ratings['class'].value_counts())
ratings.head()
```

# Задание 2

```python
keywords = pd.read_csv('ml-latest-small/keywords.csv')
keywords.head()
```

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def geo_locator(keyword):
    for region, cities in geo_data.items():
        for city in cities:
            if city in keyword:
                return region
    return 'Undefined'
```

```python
keywords['region'] = keywords['keyword'].apply(geo_locator)
print(keywords['region'].value_counts())
keywords[ keywords['region'] != 'Undefined' ].head()
```

# Задание 3

```python
movie_ratings = movies.merge(right = ratings, on = 'movieId', how = 'left', validate = '1:m')
movie_ratings.head()
```

```python
years = list(range(1950, 2011))
```

```python
def production_year(title):
    for year in years:
        if str(year) in title:
            return year
    return 1900
```

```python
movie_ratings['year'] = movie_ratings['title'].apply(production_year)
movie_ratings.head()
```

```python
movie_ratings_by_years = movie_ratings[['year', 'rating']].groupby('year').mean().sort_values(by = 'rating', ascending = False)
movie_ratings_by_years.head(30)
```

<b>Вывод:</b> Средний рейтинг с течением времени действительно падает
