```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

# Задание 1

```python
def habr_search(search_list):
    url = 'https://habr.com'
    dates = []
    titles = []
    links = []
    for search_phrase in search_list:
        params = {'q': search_phrase}
        page = requests.get(url + '/ru/search', params)
        soup = BeautifulSoup(page.text, 'html.parser')        
        for article in soup.find_all('div', {'class': 'tm-article-snippet'}):
            dates.append(article.find('a', {'class': 'tm-article-datetime-published'}).find('time').text)
            link = article.find('a', {'class': 'tm-title__link'})
            titles.append(link.find('span').text)        
            links.append(url + link.get('href'))

    return pd.DataFrame({'Date': dates, 'Title': titles, 'Link': links}).drop_duplicates()
```

```python
df = habr_search(['python', 'errbot', 'анализ данных'])
df
```

# Задание 2

```python
def habr_search_ext(search_list, pages = 1):
    url = 'https://habr.com'
    dates = []
    titles = []
    links = []
    texts = []
    likes = []
    for page_num in range(1, pages + 1):        
        for search_phrase in search_list:            
            params = {'q': search_phrase}
            page = requests.get(url + '/ru/search/page' + str(page_num), params)
            time.sleep(0.3)
            soup = BeautifulSoup(page.text, 'html.parser')

            for like in soup.find_all('span', {'class': 'tm-votes-meter__value_rating'}):
                likes.append(like.text)
            
            for article in soup.find_all('div', {'class': 'tm-article-snippet'}):
                dates.append(article.find('a', {'class': 'tm-article-datetime-published'}).find('time').text)
                link = article.find('a', {'class': 'tm-title__link'})
                titles.append(link.find('span').text)
                text_link = url + link.get('href')
                links.append(text_link)
                text_page = requests.get(text_link)
                if text_page.status_code == 200:
                    soup = BeautifulSoup(text_page.text, 'html.parser') 
                    texts.append(soup.find('div', {'class': 'article-formatted-body'}).text.replace('\n', ''))
                else:
                    texts.append('')

    return pd.DataFrame({'Date': dates, 'Title': titles, 'Link': links, 'Text': texts, 'Likes': likes}).drop_duplicates()
```

```python
df_ext = habr_search_ext(['python', 'анализ данных'], 2)
df_ext
```
